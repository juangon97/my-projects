This is my attempt of trying to model the Lorenz Attractor. 

$$
\frac{dx}{dt} = \sigma (y-x)
$$

$$
\frac{dy}{dt} = x(\rho - z) - y
$$

$$
\frac{dz}{dt} = xy - \beta y
$$


```python
import matplotlib.pyplot as plt
import math as math
```

```python
sigma = 10
rho = 28
beta = 8/3
```

```python
dt = 0.00000001
y = float(0.000001)
x = float(0.000001)
z = float(0.000001)
y_arr = []
x_arr = []
z_arr = []

for i in range(0,100000):
    dx = -sigma * (y - x) * dt
    dy = x * (rho - z) - y * dt
    dz = x * y - beta * z * dt              
    x = x + dx
    y = y + dy
    z = z + dz
        
    x_arr.append(x)
    y_arr.append(y)
    z_arr.append(z)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=True,figsize=(15,7))
ax = axs[0]
ax.plot(y_arr,x_arr, 'o',markersize=0.0099)
ax.set_title('x-y phase')

ax = axs[2]
plt.plot(z_arr,x_arr,'o',markersize=0.0099)
ax.set_title('x-z phase')
ax.get_yaxis().set_visible(False)

ax = axs[1]
ax.plot(y_arr,z_arr, 'o',markersize=0.0099)
ax.set_title('y-z phase')
ax.get_yaxis().set_visible(False)
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(18,8))
ax = plt.axes(projection='3d')
ax.scatter3D(x_arr, y_arr, z_arr, color='grey',alpha = 0.3)
ax.set_axis_off()
ax.set_title('Lorenz Attractor')
```

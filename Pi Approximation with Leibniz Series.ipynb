###### The denominator is composed of the next odd number. Thus, this shows that the equation converges to $\pi$ divided by 4

$$
1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} - ... = \frac{\pi}{4}
$$

```python
import numpy as np
import matplotlib.pyplot as plt 
π = np.pi
π / 4
```

```python
def leibniz(terms):
    π = 0
    for i in range(terms):
        denominator = 1
        numerator = (i*2 + 3)
        term = denominator / numerator
        if i%2:
            π -= term
        else:
            π += term
    return(π)
    
```

```python
π_arr = []
odd_arr = []
n_terms = 150
y_arr = []

for i in range(0,n_terms):
    π_approx = 1 - leibniz(i)
    odd_terms = i*2 + 3
    y = π / 4
    
    π_arr.append(π_approx)
    odd_arr.append(odd_terms)
    y_arr.append(y)
```

```python
plt.figure(figsize=(16,8))
plt.plot(odd_arr,π_arr,'o-',label='Leibniz Series')
plt.plot(odd_arr,y_arr,'--',color='red',label='π/4')
plt.xlim(0)
plt.grid(alpha=0.3)
plt.legend(fontsize=14)
```
